```python
from relic.graphs.graphs import get_precision_recall #, draw_graph
import pandas as pd
import networkx as nx
import os
import glob

import numpy as np
import pickle
import seaborn as sns
import logging
from tqdm.auto import tqdm
```

```python
# Input Directory Settings

base_dir = '/Users/suhail/mnt/roscoe/data/relic/matfreq2/'
```

```python
relic_result_file = 'relic_inferred_graph.csv'

def result_dict_to_df(result_dict, nb_name='nb', result_type='relic'):
    result_dict.update({'nb_name': nb_name, 
                       'result_type': result_type})
    return pd.DataFrame({k: [v] for k, v in result_dict.items()})

def process_directory(base_dir):
    workflow_ids = sorted([d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))])
    all_results = []
    all_gt_graphs = {}
    edges_df = pd.DataFrame()
    relic_result_df = pd.DataFrame()
    
    for i in tqdm(workflow_ids):
        try:
            gt_file = f"{base_dir}/{i}/{i}_gt_fixed.pkl"
            gt_graph = nx.read_gpickle(gt_file)
            all_gt_graphs[i] = gt_graph
            
            inferred_dir = f"{base_dir}/{i}/inferred/"
                       
            relic_graph_file = f"{inferred_dir}/{relic_result_file}"
            relic_graph_result = nx.read_edgelist(relic_graph_file)
            result_dict = get_precision_recall(gt_graph, relic_graph_result)
            relic_result_df = relic_result_df.append(result_dict_to_df(result_dict, nb_name=i), ignore_index=True)
            
        except FileNotFoundError as e:
            tqdm.write(str(e))
            relic_result_df = relic_result_df.append(pd.DataFrame({'nb_name': [i]}))
                        
    return relic_result_df.sort_values('nb_name'), all_gt_graphs
```

```python
r_df, all_gt_graphs = process_directory(base_dir)
```
